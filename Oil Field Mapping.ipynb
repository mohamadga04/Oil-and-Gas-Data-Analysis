```python
# Import important libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set the plot style to 'default'
plt.style.use('default')
# Set figure size
fig = plt.figure(figsize=(8,6))
```

```python
# Import file
df = pd.read_csv('demoxy.txt', sep='\t')

# Get X, Y As List
x = df['XCoordinate']
y = df['YCoordinate']
Z = df['TotalDepth']
df = df.fillna(0)  # Replaced 'NaN' with 0
```

```python
df.head()
```

```python
# Plot Triangular Color Filled Contour
plt.tricontourf(x, y, Z, cmap='gist_earth')
plt.colorbar()
plt.tricontour(x, y, Z)

# Set Well Shapes
plt.scatter(x, y, color='black', marker='^')

plt.xlabel("X Coordinate")  # Set X label
plt.ylabel("Y Coordinate")  # Set Y label
plt.title("Limestone Top", fontdict=({'size': 16}))  # Set plot title
```
