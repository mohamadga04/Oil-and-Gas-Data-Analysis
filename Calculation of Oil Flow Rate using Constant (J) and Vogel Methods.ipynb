```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Constants
J_constant = 0.7  # Productivity Index(J)constant value in STB/psi
Pr = 2800  # Reservoir pressure in psi

def calculate_Qo(Pwf_values):
    """Calculate Qo for a list of Pwf values."""
    Qo_values = [J_constant * (Pr - Pwf) for Pwf in Pwf_values]
    return Qo_values

# Example: Input a list of Pwf values
Pwf_values = [2800, 2600, 2400, 2200, 2000, 1800, 1600, 1400, 1200, 1000, 800, 600, 400, 200, 0]
Qo_result = calculate_Qo(Pwf_values)

# IPR Plot using J Constant Method
plt.plot(Qo_result, Pwf_values, marker='o', linestyle='-')
plt.xlabel('Qo (STB/Day)')
plt.ylabel('Pwf (psi)')
plt.title('IPR Plot using J Constant Method')
plt.grid(True)
plt.show()
```

```python
# Create a DataFrame for the results
result_df1 = pd.DataFrame({'Pwf (psi)': Pwf_values, 'Qo (STB/Day)': Qo_result})

# Save the DataFrame to an Excel file
excel_filename = 'oil_flow_results.xlsx'
result_df1.to_excel(excel_filename, index=False)

# Display the DataFrame
print(result_df1)

print(f"\nResults have been saved to {excel_filename}")
```

```python
def calculate_Q(Qmax, Pr, Pwf):
    return Qmax * (1 - 0.2 * (Pwf / Pr) - 0.8 * ((Pwf / Pr) ** 2))

# Given values
Qmax_given = 1203.508771
Pr = 2800  # psi
Pwf_values = [2800, 2600, 2400, 2200, 2000, 1800, 1600, 1400, 1200, 1000, 800, 600, 400, 200, 0]

# Step 2: Calculate Q for each Pwf value using the given Qmax
Q_values = [calculate_Q(Qmax_given, Pr, Pwf) for Pwf in Pwf_values]

# Create a DataFrame for the results
result_df = pd.DataFrame({'Pwf (psi)': Pwf_values, 'Qo (STB/Day)': Q_values})

# IPR Plot using Vogel Method
plt.figure(figsize=(10, 6))
plt.plot(result_df['Qo (STB/Day)'], result_df['Pwf (psi)'], marker='o')

plt.title('IPR Plot using Vogel Method')
plt.xlabel('Pwf (psi)')
plt.ylabel('Qo (STB/Day)')
plt.grid(True)
plt.show()
```

```python
# Save the DataFrame to an Excel file
excel_filename = 'Vogel_results.xlsx'
result_df.to_excel(excel_filename, index=False)

# Display the DataFrame
print(result_df)
print(f"\nResults have been saved to {excel_filename}")
```
